#Merging Dataframe

```python
df1 = pd.DataFrame({'S_Name' : ['Mohit', 'Jatin', 'Prateek', 'Mohit'],
                   'CGPA' : [2, 3, 5, 8]})
df1
```

```python
df2 = pd.DataFrame({'T_Name': ["Mohit", "Jatin", "Prateek", "Mohit"],
                    "CGPA" : [9, 8, 6, 7]})
```

```python
df2
```

```python
df1.merge(df2, how = 'inner')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how = 'outer')
```
